```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

runs_path = '/home/ubuntu/snake-rl/runs'
os.chdir(runs_path)
os.listdir()
```

```python
run = 0
run_path=os.path.join(runs_path,os.listdir()[0])

Y_FILENAME = os.path.join(run_path,'T.npy')
y = np.load(Y_FILENAME)
y = pd.Series(y,name='Durations')
y.index.name='episodes'

Z_FILENAME = os.path.join(run_path,'G.npy')
z = np.load(Z_FILENAME)
z = pd.Series(z,name='Returns')
z.index.name='episodes'

df = pd.DataFrame([y,z]).transpose()
```

```python
window_size=1

fig,axs=plt.subplots(2,1,figsize=[20,10])
df['Durations'].rolling(window_size).mean().plot(ax=axs[0]);
df['Returns'].rolling(window_size).mean().plot(ax=axs[1],color='red');
```
